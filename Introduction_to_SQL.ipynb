#**Question 1 : Explain the fundamental differences between DDL, DML, and DQL commands in SQL. Provide one example for each type of command.**

**Answer :**

SQL commands are categorized based on the type of operations they perform on the database.

**1. DDL (Data Definition Language)**

DDL commands are used to define, modify, or delete the structure of database objects such as tables, schemas, and indexes.

Examples of DDL commands:

CREATE

ALTER

DROP

TRUNCATE


**2. DML (Data Manipulation Language)**

DML commands are used to manipulate data stored in database tables.

Examples of DML commands:

INSERT

UPDATE

DELETE


**3. DQL (Data Query Language)**

DQL commands are used to retrieve data from the database.

Example of DQL command:

SELECT


````python
import sqlite3

# Connect to an in-memory SQLite database
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

print("Connected to an in-memory SQLite database.")
````

### Execute DDL Command (CREATE TABLE)

First, let's create the `Students` table using the DDL `CREATE TABLE` command.

````python
ddl_command = """
CREATE TABLE Students (
    StudentID INT PRIMARY KEY,
    StudentName VARCHAR(50)
);
"""
cursor.execute(ddl_command)
conn.commit()
print("Table 'Students' created successfully.")
````

### Execute DML Command (INSERT INTO)

Next, we'll insert a record into the `Students` table using the DML `INSERT INTO` command.

````python
dml_command = "INSERT INTO Students VALUES (1, 'Shreya');"
cursor.execute(dml_command)
conn.commit()
print("Data inserted successfully.")
````

### Execute DQL Command (SELECT)

Finally, let's retrieve the data from the `Students` table using the DQL `SELECT` command.

````python
dql_command = "SELECT * FROM Students;"
cursor.execute(dql_command)
results = cursor.fetchall()

print("Query Results:")
for row in results:
    print(row)

# Close the connection (optional for in-memory, but good practice)
conn.close()
print("\nDatabase connection closed.")
````

#**Question 4 : What is a Common Table Expression (CTE) in SQL, and what are its main benefits? Provide a simple SQL example demonstrating its usage.**

**Answer :**

A CTE (Common Table Expression) is a temporary named result set defined using the WITH keyword.

Benefits:

Improves query readability

Simplifies complex queries

Can be reused within the same query


````python
import sqlite3

# Connect to a new in-memory SQLite database for the CTE example
conn_cte = sqlite3.connect(':memory:')
cursor_cte = conn_cte.cursor()

print("Connected to a new in-memory SQLite database for CTE example.")
````

### Create the `Orders` Table and Insert Sample Data

````python
create_table_sql = """
CREATE TABLE Orders (
    OrderID INT PRIMARY KEY,
    Region VARCHAR(50),
    SalesAmount DECIMAL(10, 2)
);
"""
cursor_cte.execute(create_table_sql)
conn_cte.commit()
print("Table 'Orders' created successfully.")

insert_data_sql = """
INSERT INTO Orders (OrderID, Region, SalesAmount) VALUES
(1, 'North', 50000.00),
(2, 'South', 120000.00),
(3, 'East', 80000.00),
(4, 'West', 150000.00),
(5, 'North', 70000.00),
(6, 'South', 90000.00),
(7, 'East', 110000.00),
(8, 'West', 60000.00);
"""
cursor_cte.execute(insert_data_sql)
conn_cte.commit()
print("Sample data inserted into 'Orders' table.")
````

### Execute the CTE Query

````python
cte_query = """
WITH RegionalSales AS (
    SELECT
        Region,
        SUM(SalesAmount) AS TotalSales
    FROM
        Orders
    GROUP BY
        Region
)
SELECT
    Region,
    TotalSales
FROM
    RegionalSales
WHERE
    TotalSales > 100000;
"""

cursor_cte.execute(cte_query)
cte_results = cursor_cte.fetchall()

print("CTE Query Results (Regions with Total Sales > 100,000):")
for row in cte_results:
    print(row)

# Close the connection
conn_cte.close()
print("\nDatabase connection for CTE example closed.")
````

#**Question 9 : Display top 2 most expensive products in each category using CTE and Window Function.**

````python
top_products_query = """
WITH RankedProducts AS (
    SELECT
        ProductID,
        ProductName,
        Category,
        Price,
        ROW_NUMBER() OVER (PARTITION BY Category ORDER BY Price DESC) as rn
    FROM
        Products
)
SELECT
    ProductID,
    ProductName,
    Category,
    Price
FROM
    RankedProducts
WHERE
    rn <= 2
ORDER BY
    Category, Price DESC;
"""

cursor_ecommerce.execute(top_products_query)
top_products_results = cursor_ecommerce.fetchall()

print("Top 2 Most Expensive Products per Category:")
print("-------------------------------------------")
for row in top_products_results:
    print(f"Category: {row[2]}, Name: {row[1]}, Price: {row[3]}, ID: {row[0]}")

# The connection is kept open for further queries if needed.
````

### Which 3 movie categories have the highest rental counts?

To identify the top 3 movie categories with the highest rental counts, we would need to join the `category`, `film_category`, `inventory`, and `rental` tables. The query would group the results by category name and count the rentals for each category.

**Assumed Sakila Schema Tables and Relationships:**
*   `category`: `category_id`, `name`
*   `film_category`: `film_id`, `category_id` (links `film` to `category`)
*   `inventory`: `inventory_id`, `film_id`, `store_id`
*   `rental`: `rental_id`, `inventory_id`

**SQL Query:**

```sql
SELECT
    c.name AS category_name,
    COUNT(r.rental_id) AS rental_count
FROM
    category c
JOIN
    film_category fc ON c.category_id = fc.category_id
JOIN
    inventory i ON fc.film_id = i.film_id
JOIN
    rental r ON i.inventory_id = r.inventory_id
GROUP BY
    c.name
ORDER BY
    rental_count DESC
LIMIT 3;
```

**Explanation:**
1.  **`FROM category c`**: Starts with the `category` table to get category names.
2.  **`JOIN film_category fc ON c.category_id = fc.category_id`**: Links categories to films.
3.  **`JOIN inventory i ON fc.film_id = i.film_id`**: Links films to inventory items, which represent physical copies available for rent.
4.  **`JOIN rental r ON i.inventory_id = r.inventory_id`**: Links inventory items to actual rental events.
5.  **`GROUP BY c.name`**: Groups the results by each category name.
6.  **`COUNT(r.rental_id)`**: Counts the number of rentals for each grouped category.
7.  **`ORDER BY rental_count DESC`**: Sorts the categories in descending order based on their rental count.
8.  **`LIMIT 3`**: Restricts the output to the top 3 categories.

Provide the SQL query and explanation for calculating how many films are available at each store and how many have never been rented, assuming a full Sakila schema (film, inventory, rental, store).


## SQL Query for Films Available and Never Rented

```sql
SELECT
    s.store_id,
    COUNT(DISTINCT i.film_id) AS films_available_at_store,
    SUM(CASE WHEN r.rental_id IS NULL THEN 1 ELSE 0 END) AS films_never_rented
FROM
    store s
JOIN
    inventory i ON s.store_id = i.store_id
LEFT JOIN
    rental r ON i.inventory_id = r.inventory_id
GROUP BY
    s.store_id
ORDER BY
    s.store_id;
```


### Explanation of the SQL Query:

This SQL query is designed to calculate two metrics for each store: the total number of unique films available and the number of those films that have never been rented.

1.  **`FROM store s`**: We start by selecting from the `store` table, aliased as `s`, as we want to aggregate results per store.

2.  **`JOIN inventory i ON s.store_id = i.store_id`**: We join with the `inventory` table, aliased as `i`. The `inventory` table lists all the copies of films available in each store. This join links each store to its available film inventory.

3.  **`LEFT JOIN rental r ON i.inventory_id = r.inventory_id`**: A `LEFT JOIN` is used with the `rental` table, aliased as `r`. This is crucial because we want to include all inventory items (films) from each store, even if they have never been rented. If an `inventory_id` from the `inventory` table has no matching record in the `rental` table, the columns from the `rental` table will be `NULL` for that row. This allows us to identify unrented films.

4.  **`GROUP BY s.store_id`**: The results are grouped by `store_id` to perform aggregations (counts and sums) for each individual store.

5.  **`SELECT s.store_id, COUNT(DISTINCT i.film_id) AS films_available_at_store, SUM(CASE WHEN r.rental_id IS NULL THEN 1 ELSE 0 END) AS films_never_rented`**:
    *   **`s.store_id`**: Selects the ID of the store.
    *   **`COUNT(DISTINCT i.film_id) AS films_available_at_store`**: Counts the number of unique `film_id`s in the `inventory` for each store. This represents the total unique films stocked at that store.
    *   **`SUM(CASE WHEN r.rental_id IS NULL THEN 1 ELSE 0 END) AS films_never_rented`**: This is a conditional sum. For each row in the joined result, if `r.rental_id` is `NULL` (meaning the particular inventory item has never been rented), it counts as `1`. Otherwise, it counts as `0`. The `SUM` then tallies up all inventory items that have never been rented for each store. This accurately identifies films that have never had a rental record associated with them.

6.  **`ORDER BY s.store_id`**: Orders the final result set by store ID for better readability.

   - `AS payment_month`: This assigns an alias `payment_month` to the extracted year-month string, making the output column more readable.
   - `SUM(amount)`: This aggregate function calculates the sum of all `amount` values for each group (which will be each month).
   - `AS total_revenue`: This assigns an alias `total_revenue` to the sum of amounts.

   - This specifies that the query will retrieve data from the `payment` table.

   - `= '2023'`: This condition filters the records to include only those payments made in the year 2023. This is crucial for isolating the data to the requested year.

**4. `GROUP BY payment_month`**
   - This clause groups the rows based on the `payment_month` (the 'YYYY-MM' string). All payments occurring in the same month will be grouped together, allowing the `SUM(amount)` function to calculate the total revenue for that specific month.

**5. `ORDER BY payment_month`**
   - This clause sorts the final result set in ascending order by `payment_month`, ensuring that the monthly revenues are displayed chronologically from January to December.

## Explain Sub-Question 5: Customers with More Than 10 Rentals in the Last 6 Months

### Subtask:
Provide the SQL query and explanation for identifying customers who have rented more than 10 times in the last 6 months, assuming a full Sakila schema (customer, rental).

### Assumed Sakila Schema Tables and Relationships:
*   `customer`: `customer_id`, `first_name`, `last_name`, `email`
*   `rental`: `rental_id`, `customer_id`, `rental_date`

### SQL Query:

```sql
SELECT
    c.first_name,
    c.last_name,
    c.email,
    COUNT(r.rental_id) AS rental_count
FROM
    customer c
JOIN
    rental r ON c.customer_id = r.customer_id
WHERE
    r.rental_date >= date('now', '-6 months') -- Filters rentals from the last 6 months
GROUP BY
    c.customer_id, c.first_name, c.last_name, c.email
HAVING
    COUNT(r.rental_id) > 10
ORDER BY
    rental_count DESC;
```

### Explanation of the SQL Query:

This SQL query identifies customers who have made more than 10 rentals within the last six months.

1.  **`FROM customer c`**: We start by selecting from the `customer` table, aliased as `c`, to get customer details.

2.  **`JOIN rental r ON c.customer_id = r.customer_id`**: We join the `customer` table with the `rental` table, aliased as `r`, on `customer_id`. This links each customer to their rental records.

3.  **`WHERE r.rental_date >= date('now', '-6 months')`**: This `WHERE` clause filters the rental records to include only those that occurred within the last six months from the current date.
    *   `date('now')` gets the current date.
    *   `'-6 months'` subtracts six months from the current date.

4.  **`GROUP BY c.customer_id, c.first_name, c.last_name, c.email`**: The results are grouped by customer to count the total rentals for each individual customer.

5.  **`SELECT c.first_name, c.last_name, c.email, COUNT(r.rental_id) AS rental_count`**:
    *   Selects the customer's first name, last name, and email.
    *   `COUNT(r.rental_id)` calculates the total number of rentals for each customer within the filtered time period.
    *   `AS rental_count` assigns an alias to the count column for readability.

6.  **`HAVING COUNT(r.rental_id) > 10`**: This `HAVING` clause filters the grouped results, keeping only those customers who have a `rental_count` greater than 10. This is crucial for fulfilling the condition of "more than 10 times."

7.  **`ORDER BY rental_count DESC`**: Finally, the results are ordered in descending order based on the `rental_count`, so customers with the highest number of rentals appear first.

## SQL Query for Customers with More Than 10 Rentals in Last 6 Months

```sql
SELECT
    c.first_name,
    c.last_name,
    c.email,
    COUNT(r.rental_id) AS total_rentals
FROM
    customer c
JOIN
    rental r ON c.customer_id = r.customer_id
WHERE
    r.rental_date >= date('now', '-6 months')
GROUP BY
    c.customer_id, c.first_name, c.last_name, c.email
HAVING
    COUNT(r.rental_id) > 10
ORDER BY
    total_rentals DESC;
```

### Explanation of the SQL Query:

This SQL query is designed to identify customers who have made more than 10 rentals within the last six months.

1.  **`SELECT c.first_name, c.last_name, c.email, COUNT(r.rental_id) AS total_rentals`**
    *   `c.first_name`, `c.last_name`, `c.email`: These select the customer's first name, last name, and email address to identify them in the report.
    *   `COUNT(r.rental_id) AS total_rentals`: This counts the number of rental records (`rental_id`) for each customer. The result is aliased as `total_rentals` for clarity.

2.  **`FROM customer c JOIN rental r ON c.customer_id = r.customer_id`**
    *   `FROM customer c`: The query starts by selecting from the `customer` table, aliased as `c`.
    *   `JOIN rental r ON c.customer_id = r.customer_id`: This joins the `customer` table with the `rental` table (aliased as `r`) using `customer_id`. This linkage is essential to associate rentals with specific customers.

3.  **`WHERE r.rental_date >= date('now', '-6 months')`**
    *   `WHERE r.rental_date >= date('now', '-6 months')`: This is the filtering condition. It restricts the rental records to only those that occurred within the last six months from the current date. `date('now', '-6 months')` is a SQLite function that calculates the date exactly six months ago.

4.  **`GROUP BY c.customer_id, c.first_name, c.last_name, c.email`**
    *   `GROUP BY`: This clause groups the rows by each unique customer. All aggregate functions (like `COUNT()`) will then operate on these groups. Including `first_name`, `last_name`, and `email` in the `GROUP BY` clause is standard practice when these columns are selected without being part of an aggregate function.

5.  **`HAVING COUNT(r.rental_id) > 10`**
    *   `HAVING`: This clause filters the *grouped* results. It ensures that only those customer groups (i.e., individual customers) whose `total_rentals` count is greater than 10 are included in the final output.

6.  **`ORDER BY total_rentals DESC`**
    *   `ORDER BY total_rentals DESC`: This sorts the final result set in descending order based on the `total_rentals` count, showing the customers with the highest number of rentals at the top.
